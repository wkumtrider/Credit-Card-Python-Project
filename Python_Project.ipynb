```python
import sqlite3

# create the sqlite DB
db = sqlite3.connect("CreditCard.db")

# read CSV into dataframe
df1 = pd.read_csv("Jan 2019.csv")

# convert the pandas dataframe to a sqlite table
df1.to_sql("purchases", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)

os.getcwd()
```

```python
# view dataframe
df1
```

```python
query = """
SELECT * 
FROM purchases LIMIT 10;
"""
run_query(query)
```

query = """
SELECT COUNT(DISTINCT 6)
FROM purchases
"""
run_query(query)

```python
query = """
SELECT SUM(Amount)AS total_amount FROM purchases
"""
run_query(query)
```

```python
query = """
SELECT MIN(Amount) AS Min_cost, MAX(Amount) AS Max_Amount FROM purchases
"""
run_query(query)
```

```python
query = """
SELECT *, ABS(Amount) FROM purchases
"""
run_query(query)
```

```python
query = """
SELECT SUM(ABS(Amount))FROM purchases
"""
run_query(query)
```

```python
# Group category by type and show count of each category type
query = """
SELECT COUNT(category), category
FROM purchases GROUP BY category
ORDER BY COUNT(Category)DESC
"""
run_query(query)
```

```python
# Show total cost of each category
query = """
SELECT Count(category), category, SUM(ABS(Amount)) AS Total_Spend
FROM purchases GROUP BY category
ORDER BY COUNT(Category)DESC
"""
run_query(query)
```
