```python
import sqlite3

# create the sqlite DB
db = sqlite3.connect("CreditCard.db")

# read CSV into dataframe
df1 = pd.read_csv("Jan 2019.csv")

# convert the pandas dataframe to a sqlite table
df1.to_sql("purchases", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)

os.getcwd()
```

```python
# view dataframe
df1
```

```python
query = """
SELECT * 
FROM sqlite_master
ORDER BY name;
"""
run_query(query)
```

```python
query = """
SELECT * 
FROM purchases LIMIT 10;
"""
run_query(query)
```

query = """
SELECT COUNT(DISTINCT 6)
FROM purchases
"""
run_query(query)

```python
query = """
SELECT COUNT(*) AS category_count FROM purchases
GROUP BY category
"""
run_query(query)
```

```python
query = """
SELECT SUM(Amount)AS total_amount FROM purchases
"""
run_query(query)
```

```python
query = """
SELECT MIN(Amount) AS Min_cost FROM purchases
"""
run_query(query)
```

```python
query = """
SELECT category FROM purchases
ORDER BY category DESC LIMIT 5
"""
run_query(query)
```
